```python
# Linear Regression with Scikit-Learn
import numpy as np
from sklearn.linear_model import LinearRegression
x=np.arange(4).reshape(-1,1)
y=np.array([2,7,3,1])
model=LinearRegression().fit(x,y)
corr=model.score(x,y)
print("coefficient determination:",corr)
y_pred=model.predict(x)
print("predicted Response:",y_pred)
```

```python
# Multiple Linear Regression with Scikit-learn
import numpy as np
from sklearn.linear_model import LinearRegression
x=np.array([[1,3],[2,8],[3,10],[4,9]])
y=np.arange(4)
model=LinearRegression().fit(x,y)
corr=model.score(x,y)
print("coefficient determination:",corr)
print("Intercept:",model.intercept_)
print("slope:",model.coef_)
y_pred=model.predict(x)
print("predicted Response:",y_pred)
```

```python
# Calculating Correlation Coefficient with NumPy
import numpy as np
x=np.arange(6)
y=np.array([1,7,3,9,2,4])
corr=np.corrcoef(x,y)
print("Correlation coefficient:",corr)
```

```python
# Calculating Pearson Correlation Coefficient using SciPy  
from scipy import stats
import numpy as np
x=np.arange(4)
y=np.array([1,7,2,8])
corr=stats.pearsonr(x,y)
print("pearson coefficient:",corr)
```

```python
# Plotting the PDF of a Normal Distribution 
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
data=np.arange(1,10,0.01)
pdf=norm.pdf(data,loc=5.3,scale=1)
sb.set_style('whitegrid')
sb.lineplot(x=data,y=pdf,color='black')

plt.xlabel('heights')
plt.ylabel('probability distribution')
```

```python
# Correlation and Linear Regression Plot 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
y=pd.Series([1,2,3,4,3,5,4])
x=pd.Series([1,2,3,4,5,6,7])
corr=y.corr(x)
print("correlation:",corr)
plt.scatter(x,y)
plt.plot(np.unique(x),np.poly1d(np.polyfit(x,y,1))(np.unique(x)),color='red')
plt.xlabel('x-value')
plt.ylabel('y-value')
plt.show()
```

```python
# Frequency and Conditional Frequency Distribution
import numpy as np
import pandas as pd
import nltk
items=['apple','apple','kiwi','cabbage','cabbage','potato']
nltk.FreqDist(items)
c_items=[('f','apples'),('f','apples'),('f','kiwi'),('v','cabbage'),('v','cabbage'),('v','potato')]
cfd=nltk.ConditionalFreqDist(c_items)
cfd.conditions()
cfd.plot()
cfd['v']
```

```python
# Exploring Conditional Frequency Distribution with Brown Corpus
import nltk
nltk.download('brown')
from nltk.corpus import brown
cfd=nltk.ConditionalFreqDist([(genre,word) for genre in brown.categories() for word in brown.words(categories=genre)])
cfd.conditions()
cfd.tabulate(conditions=['government','humor','reviews'],samples=['leadership','worship','hardship'])
cfd.plot(conditions=['government','humor','reviews'],samples=['leadership','worship','hardship'])
cfd.tabulate(conditions=['government','humor','reviews'],samples=['leadership','worship','hardship'],cumulative='Trure')
news_fd=cfd['news']
news_fd.most_common(3)
news_fd['the']
```

```python
# Exploring Conditional Frequency Distribution with NLTK's Names Corpus
import nltk
from nltk.corpus import names 
nltk.download('names')
nt=[(fid.split('.')[0],name[-1]) for fid in names.fileids() for name in names.words(fid)]
cfd2=nltk.ConditionalFreqDist(nt)
cfd2['female']['a']
cfd2['male']['a']
cfd2['female']>cfd2['male']
cfd2.tabulate(samples=['a','e'])
cfd2.plot()
```

```python
# Calculating the Average of a List
def average(num):
    sum_n=0
    for i in num:
        sum_n=sum_n+i
    avg=sum_n/len(num)
    return avg
average([1,2,3])
```
